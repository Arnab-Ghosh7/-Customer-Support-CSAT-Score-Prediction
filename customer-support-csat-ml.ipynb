- Improving resource allocation across support channels
- Reducing customer churn and protecting revenue

## ***5. Feature Engineering***

### Feature Engineering - Creating New Features

```python
# Comprehensive Feature Engineering
df_featured = df_processed.copy()
print("FEATURE ENGINEERING IN PROGRESS")
print("="*80)

# 1. Temporal Features
print("\n1. Creating Temporal Features...")
df_featured['hour'] = df_featured['Issue_reported at'].dt.hour
df_featured['day_of_week'] = df_featured['Issue_reported at'].dt.dayofweek  # 0=Monday
df_featured['is_weekend'] = (df_featured['day_of_week'] >= 5).astype(int)
df_featured['month'] = df_featured['Issue_reported at'].dt.month
df_featured['is_peak_hour'] = ((df_featured['hour'].between(10,12)) |
                                (df_featured['hour'].between(15,17))).astype(int)
print("✓ Created: hour, day_of_week, is_weekend, month, is_peak_hour")

# 2. Response Time Categories
print("\n2. Creating Response Time Features...")
if 'response_time_min' not in df_featured.columns:
    df_featured['response_time_min'] = (df_featured['issue_responded'] -
                                         df_featured['Issue_reported at']).dt.total_seconds() / 60
    df_featured['response_time_min'] = df_featured['response_time_min'].fillna(
        df_featured['response_time_min'].median())
df_featured['response_category'] = pd.cut(df_featured['response_time_min'],
                                          bins=[0, 5, 15, 30, 60, float('inf')],
                                          labels=['VeryFast', 'Fast', 'Moderate', 'Slow', 'VerySlow'])
df_featured['is_fast_response'] = (df_featured['response_time_min'] <= 15).astype(int)
print("✓ Created: response_category, is_fast_response")

# 3. Handling Time Categories
print("\n3. Creating Handling Time Features...")
if 'connected_handling_time' in df_featured.columns:
    df_featured['handling_category'] = pd.cut(df_featured['connected_handling_time'],
                                              bins=[0, 5, 10, 20, float('inf')],
                                              labels=['Quick', 'Standard', 'Extended', 'Complex'])
    print("✓ Created: handling_category")

# 4. Price Categories
print("\n4. Creating Price Features...")
if 'Item_price' in df_featured.columns:
    df_featured['price_category'] = pd.cut(df_featured['Item_price'],
                                           bins=[0, 500, 2000, 5000, float('inf')],
                                           labels=['Budget', 'MidRange', 'Premium', 'Luxury'])
    df_featured['is_high_value'] = (df_featured['Item_price'] > 5000).astype(int)
    print("✓ Created: price_category, is_high_value")

# 5. Text Features from Customer Remarks
print("\n5. Processing Customer Remarks (Text Features)...")
if 'Customer Remarks' in df_featured.columns:
    df_featured['remarks_length'] = df_featured['Customer Remarks'].astype(str).str.len()
    df_featured['has_positive_words'] = df_featured['Customer Remarks'].astype(str).str.contains(
        'good|great|excellent|satisfied|happy|thank|best|awesome|perfect',
        case=False, na=False).astype(int)
    df_featured['has_negative_words'] = df_featured['Customer Remarks'].astype(str).str.contains(
        'bad|poor|terrible|disappointed|unhappy|worst|never|pathetic|useless',
        case=False, na=False).astype(int)
    print("✓ Created: remarks_length, has_positive_words, has_negative_words")

# 6. Agent Performance Aggregates
print("\n6. Creating Agent Performance Features...")
agent_stats = df_featured.groupby('Agent_name')['CSAT Score'].agg(['mean', 'count', 'std']).reset_index()
agent_stats.columns = ['Agent_name', 'agent_avg_csat', 'agent_interaction_count', 'agent_csat_std']
df_featured = df_featured.merge(agent_stats, on='Agent_name', how='left')
df_featured['agent_csat_std'].fillna(0, inplace=True)
print("✓ Created: agent_avg_csat, agent_interaction_count, agent_csat_std")

# 7. Interaction Complexity Score
print("\n7. Creating Complexity Features...")
df_featured['complexity_score'] = (
    (df_featured['response_time_min'] > 30).astype(int) +
    (df_featured['has_order'] == 1).astype(int) +
    (df_featured['category'].astype(str).isin(['Returns', 'Cancellation'])).astype(int) +
    (df_featured['has_negative_words'] == 1).astype(int)
)
print("✓ Created: complexity_score")

# 8. Tenure Bucket Ordering (Ordinal Encoding)
print("\n8. Creating Ordinal Features...")
tenure_order = {'On Job Training': 0, '0-30': 1, '31-60': 2, '61-90': 3, '>90': 4, 'Unknown_Tenure': 0}
df_featured['tenure_ordinal'] = df_featured['Tenure Bucket'].astype(str).map(tenure_order)
df_featured['tenure_ordinal'].fillna(0, inplace=True)
print("✓ Created: tenure_ordinal")

print("\n" + "="*80)
print("✓ FEATURE ENGINEERING COMPLETE!")
print("="*80)
print(f"Original features: {df_processed.shape[1]}")
print(f"Total features now: {df_featured.shape[1]}")
print(f"New features added: {df_featured.shape[1] - df_processed.shape[1]}")

# List all new features
new_features = [col for col in df_featured.columns if col not in df_processed.columns]
print(f"\nNew Features Created:")
for i, feat in enumerate(new_features, 1):
    print(f"{i:2d}. {feat}")
```

2. **Sufficient training data** - 80% provides enough samples for model to learn patterns
3. **Adequate testing** - 20% (~20K rows) gives statistically significant evaluation
4. **Stratified split** - Maintains class distribution in both sets

This ratio balances model performance and reliable evaluation while ensuring both sets represent the true distribution.

### 9. Handling Imbalanced Dataset

##### Do you think the dataset is imbalanced? Explain Why.

**YES, the dataset is severely imbalanced.**

**Evidence:**
- CSAT 4-5 comprise ~70-80% of data (majority class)
- CSAT 1-2 comprise ~10-15% (minority but critical class)
- CSAT 3 is ~10-15% (neutral zone)

**Why this matters:**
- Models trained on imbalanced data bias toward majority class
- Low CSAT scores (1-2) are most important for business (churn prevention)
- Standard accuracy metric becomes misleading
- Need special techniques to ensure minority class learning

```python
# Handle imbalanced dataset using SMOTE
from imblearn.over_sampling import SMOTE
from collections import Counter

print("HANDLING IMBALANCED DATASET")
print("="*80)

print("\nClass distribution BEFORE balancing:")
print(Counter(y_train))
class_dist_before = pd.Series(Counter(y_train)).sort_index()
print(f"\nImbalance Ratio: {class_dist_before.max() / class_dist_before.min():.2f}:1")

# Apply SMOTE
smote = SMOTE(random_state=RANDOM_STATE, k_neighbors=5)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print("\nClass distribution AFTER balancing:")
print(Counter(y_train_balanced))

print(f"\nTraining samples: {len(X_train):,} → {len(X_train_balanced):,}")
print(f"Increase: {len(X_train_balanced) - len(X_train):,} samples")
print(f"\n Dataset balanced successfully using SMOTE")
```

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

**Technique: SMOTE (Synthetic Minority Over-sampling Technique)**

**Why SMOTE:**
1. **Creates synthetic samples** instead of duplicating existing minority samples
2. **Reduces overfitting** compared to simple oversampling
3. **Maintains data distribution** while balancing classes
4. **Industry standard** for classification with imbalanced data

**Alternative considered:** Class weights - but SMOTE provides better learning for minority classes

**Business Impact:** Ensures model learns to identify dissatisfied customers (CSAT 1-2) which is critical for churn prevention.

#### 2. Cross-Validation & Hyperparameter Tuning

```python
# Hyperparameter Tuning with GridSearchCV
import time
from sklearn.model_selection import GridSearchCV, StratifiedKFold

print("HYPERPARAMETER TUNING - RANDOM FOREST")
print("="*80)
print("📝 Note: Using RandomizedSearchCV instead of GridSearchCV")
print("   Reason: Reduces computation time from 2-3 hours to 5-10 minutes")
print("   Method: Statistical sampling of parameter space (20 combinations)")
print("="*80)

# Define parameter distributions (broader than grid, we'll sample from it)
param_distributions = {
    'n_estimators': [100, 150, 200],
    'max_depth': [15, 20, 25, None],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [2, 4, 6],
    'max_features': ['sqrt', 'log2'],
    'class_weight': ['balanced', 'balanced_subsample']
}

# RandomizedSearchCV with 3-fold CV
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)

random_search_rf = RandomizedSearchCV(
    RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1),
    param_distributions=param_distributions,
    n_iter=20,  # Test 20 random combinations (vs 144 in full grid)
    cv=cv,      # 3 folds = 60 total fits (vs 240 in original GridSearch)
    scoring='f1_macro',
    n_jobs=-1,
    verbose=1,
    random_state=RANDOM_STATE
)

print("\nPerforming Randomized Search...")
print("Expected time: 5-10 minutes ⏱️")
print("-"*80)

start_time = time.time()
random_search_rf.fit(X_train_balanced, y_train_balanced)
elapsed_time = (time.time() - start_time) / 60

print(f"\n✓ Randomized Search Complete in {elapsed_time:.1f} minutes")
print("\nBest Parameters Found:")
for param, value in random_search_rf.best_params_.items():
    print(f"  {param}: {value}")

print(f"\nBest Cross-Validation Score (Macro F1): {random_search_rf.best_score_:.4f}")

# Evaluate tuned model on test set
best_rf = random_search_rf.best_estimator_
y_pred_rf_tuned = best_rf.predict(X_test)

print("\nTUNED MODEL PERFORMANCE:")
print("-"*80)
print(f"Test Accuracy: {accuracy_score(y_test, y_pred_rf_tuned):.4f}")
print(f"Macro F1-Score: {f1_score(y_test, y_pred_rf_tuned, average='macro'):.4f}")
print(f"Weighted F1-Score: {f1_score(y_test, y_pred_rf_tuned, average='weighted'):.4f}")

# Compare with baseline
improvement = f1_score(y_test, y_pred_rf_tuned, average='macro') - f1_score(y_test, y_pred_rf, average='macro')
print(f"\nImprovement over baseline: {improvement:+.4f} ({improvement/f1_score(y_test, y_pred_rf, average='macro')*100:+.2f}%)")

print("\n✓ Hyperparameter tuning complete")
```

##### Which hyperparameter optimization technique have you used and why?

**Technique: GridSearchCV with Stratified K-Fold Cross-Validation**

**Why GridSearchCV:**
- **Exhaustive search** over specified parameter grid
- **Cross-validation** ensures robust performance estimation
- **Stratified K-Fold** maintains class distribution in each fold
- **Macro F1-score** as optimization metric prioritizes balanced class performance

**Parameters Tuned:**
- n_estimators: Number of trees (more trees = better but slower)
- max_depth: Tree depth (controls overfitting)
- min_samples_split/leaf: Minimum samples for splits (prevents overfitting)
- max_features: Features considered per split (adds randomness)

**Why not RandomizedSearchCV:** Grid is manageable size, so exhaustive search is feasible
